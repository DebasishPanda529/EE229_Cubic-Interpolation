```python
import numpy as np 
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']= [20,8] 
file = open('NCub_21D070021.dat', 'r') 
samples = []

for i in file.readlines():
 samples.append(float(i)) 
#print(len(samples))
x = np.arange(1,len(samples)+1)

plt.title('Roll Number: 21D070021')
plt.subplot(1, 2, 1)
plt.plot(x, samples, 'g')

derivatives = []
for i in range(1, len(samples)-1):
 derivatives.append((samples[i+1] - samples[i-1])/2)
x_1 = np.arange(1, len(samples)-1)

derivatives_1 = []
for i in range(1, len(derivatives)-1):
 derivatives_1.append((derivatives[i+1]-derivatives[i-1])/2)
x_2 = np.arange(1, len(derivatives)-1)

plt.subplot(1, 2, 2)
plt.stem(x_2, derivatives_1, 'r')
imp_points = [118, 318, 517, 721, 917, 1118, 1317, 1521, 1717, 1921, 2117, 2321, 2517]
imp_vals = []
for i in range(0, len(imp_points)):
    imp_vals.append(samples[imp_points[i]])
# plt.stem(imp_points, imp_vals)
#print(imp_vals)
plt.show()
```
